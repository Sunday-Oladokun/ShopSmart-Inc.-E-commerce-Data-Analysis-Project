```python
# Cleaning the different IDs in our dataset- customer_id, device_id, and order_id

# functioning for shortening the id names
def handle_id(df, column):
    lst = []
    for item in df[column]:
        values = item.split('-')[-1]
        lst.append(values)
    df[column] = lst
```

```python
# Clean the columns
columns_to_modify = ['customer_id', 'device_id', 'order_id']
for column in columns_to_modify:
    handle_id(final_df, column)
```

```python
# Feature engineering to generate day, month, year, and AM/PM columns
def extract_day(date):
    
    # ensure the input is in datetime format
    date_column = pd.to_datetime(date, format='%Y-%m-%d %H:%M:%S.%f')

    # Extracting day
    day = date_column.dt.day

    return day
```

```python
# Feature engineering to generate day, month, year, and AM/PM columns
def extract_month(date):
    
    # ensure the input is in datetime format
    date_column = pd.to_datetime(date, format='%Y-%m-%d %H:%M:%S.%f')

    # Extracting month
    month = date_column.dt.month

    return month
```

```python
# Feature engineering to generate day, month, year, and AM/PM columns
def extract_year(date):
    
    # ensure the input is in datetime format
    date_column = pd.to_datetime(date, format='%Y-%m-%d %H:%M:%S.%f')

    # Extracting year
    
    year = date_column.dt.year
    
    return year
```

```python
# Feature engineering to generate day, month, year, and AM/PM columns
def extract_am_pm(date):
    
    # ensure the input is in datetime format
    date_column = pd.to_datetime(date, format='%Y-%m-%d %H:%M:%S.%f')

    # Extracting AM/PM
    am_pm = date_column.dt.strftime('%p')

    return am_pm
```

We have two columns with datetime data and we would be utilising the two as follows:

event_timestamp: This decribes the customer traffic on the website.

checked_out_at: This data contains information about the checkout time for the customer.

```python
final_df['day'] = extract_day(final_df['event_timestamp'])
final_df['month'] = extract_month(final_df['event_timestamp'])
final_df['year'] = extract_year(final_df['event_timestamp'])
final_df['am/pm'] = extract_am_pm(final_df['event_timestamp'])
```
